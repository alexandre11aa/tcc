```python
import pandas as pd
import numpy as np
import random

# Função que cálcula intensidade a partir de parâmetros da IDF

def intensidades(par, dur, tre):
    
    registro_historico = []

    for i in range(len(tre)):
        # print('')
        registro_historico.append([])

        for j in range(len(dur)):
            # print('%.2f' % (par[0] * tre[i]** par[1] / (dur[j] + par[2])**par[3]), end='\t')
            registro_historico[i].append(par[0] * tre[i]** par[1] / (dur[j] + par[2])**par[3])

    return registro_historico
```

```python
# Função que calcula porcentagem de acerto dos dados

def erro(parametros):
        
    test = intensidades(parametros,
                        [1440, 720, 600, 480, 360, 60, 30, 25, 20, 15, 10, 5],
                        [2, 5, 10, 25, 50, 75, 100])
    
    test = pd.DataFrame(test).transpose()
          
    return 1 - abs(np.mean(np.mean((treino - test) / treino)))
```
